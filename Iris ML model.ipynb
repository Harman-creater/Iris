```python
import pandas as pd                                  # for loading dataset csv file

import seaborn as sns
import matplotlib.pyplot as plt                      # for plotting the dataset

from sklearn.linear_model import LogisticRegression  # model 
from sklearn.model_selection import train_test_split # to perform train_test split
from sklearn import metrics                          # for calculating the accuracy

iris_df=pd.read_csv("iris.csv")
# print(iris_df)

# #head for first five rows
# iris_df.head()

# #if any null values there in the data
# print(iris_df.isna().sum())

sns.FacetGrid(iris_df, hue="variety", height=3).map(plt.scatter, "petal.length","sepal.width").add_legend()
#*********************************************************************

train, test = train_test_split(iris_df, test_size = 0.5)
print("Train dataset shape : ", train.shape)
print("Test dataset shape : ", test.shape)

train_X = train[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
train_Y = train.variety

test_X = test[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
test_Y = test.variety

#Using logisticRegression MODEL
model = LogisticRegression(solver='lbfgs', max_iter=150)
model.fit(train_X, train_Y)
prediction = model.predict(test_X)
print("Accuracy : ", metrics.accuracy_score(prediction,test_Y))
```
